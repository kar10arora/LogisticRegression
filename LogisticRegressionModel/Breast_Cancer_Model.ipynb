# Importing the libraries

```python
import pandas as pd
import numpy as np
```

 # DataSet

```python
df = pd.read_csv('/content/drive/MyDrive/LogisticRegressionModel/breast_cancer.csv')
df.head()
```

```python
df.info()
```

# No Null Values

```python
df.isna().sum()
```

# Split into X and Y

```python
X = df.iloc[: , 1:-1].values
Y = df.iloc[: , -1].values
```

# 2 = benign, 4 = malignant

```python
print(X)
```

```python
print(Y)
```

# Split into training and test set

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

# Model Setup

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
```

```python
classifier.fit(X_train , Y_train)
```

# Prediction

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
print(len(y_pred))
```

```python
print(len(Y_test))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test , y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy :",accuracy_score(Y_test , y_pred))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(estimator = classifier, X = X_train , y = Y_train , cv = 10)
print("ACCURACY SCORE: {:.2f} %".format(score.mean()*100))
print("STANDARD DEVIATION: {:.2f} %".format(score.std()*100))
```
